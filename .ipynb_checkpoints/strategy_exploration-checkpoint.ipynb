```python
df = pd.read_csv('data/combined_output.csv')
df.head()
```

```python
# Number of unique symbols
df['Symbol'] = df['Symbol'].str.upper()
unique_symbol_count = df['Symbol'].nunique()
print("Number of unique symbols:", unique_symbol_count)
```

```python
# Ensure the time column is datetime
df['OpenTimeFormatted'] = pd.to_datetime(df['OpenTimeFormatted'])

# Create an empty list to collect missing timestamps
missing = []

# Group by symbol
for symbol, group in df.groupby('Symbol'):
    group = group.sort_values('OpenTimeFormatted')
    
    # Create expected per-minute range
    expected_range = pd.date_range(start=group['OpenTimeFormatted'].min(),
                                   end=group['OpenTimeFormatted'].max(),
                                   freq='1min')
    
    # Find missing timestamps
    actual = set(group['OpenTimeFormatted'])
    missing_times = sorted(set(expected_range) - actual)
    
    if missing_times:
        missing.append({
            'Symbol': symbol,
            'MissingCount': len(missing_times),
            'MissingTimestamps': missing_times
        })

# Convert to DataFrame for easy viewing
missing_df = pd.DataFrame(missing)

# Show result
missing_df
```
